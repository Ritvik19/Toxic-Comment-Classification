```python
def describe(data):
    min_ = data.min()
    q1 = data.quantile(0.01)
    q25 = data.quantile(0.25)
    q50 = data.quantile(0.50)
    q75 = data.quantile(0.75)
    q99 = data.quantile(0.99)
    max_ = data.max()
    return min_, q1, q25, q50, q75, q99, max_
```

```python
import nltk, re
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer

def expand_contractions(text):
    text = re.sub(r"can't", "can not", text)
    text = re.sub(r"what's", "what is ", text)
    text = re.sub(r"'s", " ", text)
    text = re.sub(r"'ve", " have ", text)
    text = re.sub(r"n't", " not ", text)
    text = re.sub(r"i'm", "i am ", text)
    text = re.sub(r"'re", " are ", text)
    text = re.sub(r"'d", " would ", text)
    text = re.sub(r"'ll", " will ", text)
    return text

def remove_url(text):
    URL_REGEX = re.compile(r'''((http[s]?://)[^ <>'"{}|\^`[\]]*)''')
    return URL_REGEX.sub(r' ', text)

def remove_handles(text):
    HANDLES_REGEX = re.compile(r'@\S+')
    return HANDLES_REGEX.sub(r' ', text)

remove_punc = lambda x : re.sub(r"\W", ' ', x)

remove_num = lambda x : re.sub(r"\d", ' ', x)

remove_extra_spaces = lambda x : re.sub(r"\s+", ' ', x)

remove_shortwords = lambda x: ' '.join(word for word in x.split() if len(word) > 2)

lower_case = lambda x : x.lower()

with open('stopwords.txt') as f:
    sw = map(lambda x : x.strip(), f.readlines())
stop_words = set(nltk.corpus.stopwords.words('english'))|set(sw)
remove_stopwords = lambda x: ' '.join(word for word in x.split() if word not in stop_words)

ps = PorterStemmer()
ps_stem = lambda x: ' '.join(ps.stem(word) for word in x.split())

wnl = WordNetLemmatizer()
wnl_lemmatize = lambda x: ' '.join(wnl.lemmatize(word) for word in x.split())

def tag_pos(x):
    tag_list =  nltk.pos_tag(nltk.word_tokenize(x))
    pos = ""
    for t in tag_list:
        pos += t[0] +'(' + t[1] +')' + ' '
    return pos

def cleanText(x, rsw, stm, lem, tgps):
    x = str(x)
    x = remove_url(x)
    x = lower_case(x)
    x = expand_contractions(x)
    x = remove_punc(x)
    x = remove_num(x)
    x = remove_extra_spaces(x)
    x = remove_shortwords(x)
    
    if rsw:
        x = remove_stopwords(x)
    if stm:
        x = ps_stem(x)
    if lem:
        x = wnl_lemmatize(x)
    if tgps:
        x = tag_pos(x)
    return x
```

```python
import os

if not os.path.exists('data_cleaned_v2.csv'):
    data['clean_text'] = data['comment_text'].progress_apply(lambda x: cleanText(x, True, False, True, False))
    data[['clean_text', 'positive']+target].to_csv('data_cleaned_v2.csv', index=False)
else:
    data = pd.read_csv('data_cleaned_v2.csv').dropna()
```

```python
from wordcloud import WordCloud
from collections import Counter
```

```python
axes = []
word_sets = []
fig = plt.figure(figsize=(20, 24))
fig.suptitle('Most Common Words for Each Class', fontsize=36)
axes.append(plt.subplot2grid((5, 2), (0, 0), colspan=2, rowspan=2))
for i in range(6):
    axes.append(plt.subplot2grid((5, 2), (i//2+2, i%2)))

for i, col in enumerate((['positive']+target)):
    word_sets.append(Counter((' '.join(data[data[col] == 1]['clean_text'].values)).split()))
    
axes[0].imshow(WordCloud().generate_from_frequencies(
    word_sets[0]-(word_sets[1]|word_sets[2]|word_sets[3]|word_sets[4]|word_sets[5]|word_sets[6])
))    

axes[0].set_title('positive', fontsize=24)
axes[0].xaxis.set_visible(False)
axes[0].yaxis.set_visible(False)

for i, col in enumerate((target)):
    axes[i+1].imshow(WordCloud().generate_from_frequencies((word_sets[i+1]-word_sets[0])))
    axes[i+1].set_title(col, fontsize=24)
    axes[i+1].xaxis.set_visible(False)
    axes[i+1].yaxis.set_visible(False)
    
plt.savefig('WordCloud.png')    
```
