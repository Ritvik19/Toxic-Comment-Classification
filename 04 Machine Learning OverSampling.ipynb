```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from tqdm.auto import tqdm, trange
tqdm.pandas()

sns.set(style="white")
```

```python
data = pd.read_csv('data_cleaned.csv')
print(data.shape)
data.head()
```

```python
data.dropna(inplace=True)
data.reset_index(drop=True, inplace=True)
print(data.shape)
```

```python
labels= ['positive', 'identity_hate', 'insult', 'obscene', 'severe_toxic', 'threat', 'toxic']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(sublinear_tf=True)

vects = vectorizer.fit_transform(data['clean_text'])
target = data[labels].values
```

```python
from utils.ml import train_model
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
```

# Multinomial NB

```python
model = MultinomialNB()
mnb0_performance, mnb0_cm, mnb0_model = train_model(model, vects, target[:, 0], oversample='random')
print('Mean roauc', round(np.mean(mnb0_performance['roauc']), 4), '+-', round(np.std(mnb0_performance['roauc']), 4))
print('Mean f1', round(np.mean(mnb0_performance['f1']), 4), '+-', round(np.std(mnb0_performance['f1']), 4))
print('Mean accuracy', round(np.mean(mnb0_performance['accuracy']), 4), '+-', round(np.std(mnb0_performance['accuracy']), 4))
```
